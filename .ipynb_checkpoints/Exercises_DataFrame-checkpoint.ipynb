```python
import pandas as pd
import numpy as np
```

```python
pd.DataFrame(data=[None],
             index=[None],
             columns=[None])
```

```python
import matplotlib.pyplot as plt
```

```python
fig,(ax1,ax2)=plt.subplots(2, 1)

x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)
ax1.plot(x1, y1)

x2 = np.linspace(0, 4 * np.pi, 100)
y2 = np.cos(x2)
ax2.plot(x2, y2,'r')

plt.savefig('sin_cos.png')
```
